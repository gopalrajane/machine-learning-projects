```python
import pandas as pd
import numpy as np


train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
train.drop(['Alley'],axis=1,inplace=True)
test.drop(['Alley'],axis=1,inplace=True)
```

```python
import seaborn as sns

def countPlot(feature):
    sns.countplot(x=feature,data=train)
    print(train[feature].value_counts())  
```

```python
countPlot("MSSubClass")
```

```python
train.drop(['PoolQC','Fence','MiscFeature'],axis=1,inplace=True)
test.drop(['PoolQC','Fence','MiscFeature'],axis=1,inplace=True)
```

```python
test['MasVnrType'] = test['MasVnrType'].fillna("None")
train['MasVnrType'] = train['MasVnrType'].fillna("None")
```

```python
print(train["LotFrontage"].mean())
print(train["LotFrontage"].median())

print(test["LotFrontage"].mean())
print(test["LotFrontage"].median())
```

```python
train['LotFrontage_mean'] = train['LotFrontage'].fillna(train['LotFrontage'].mean())
train['LotFrontage_median'] = train['LotFrontage'].fillna(train['LotFrontage'].median())
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')

train[['SaleType']] = imputer.fit_transform(train[['SaleType']])
test[['SaleType']] = imputer.transform(test[['SaleType']])
```

```python
imputer3 = SimpleImputer(strategy='most_frequent')


test[['MSZoning']] = imputer3.fit_transform(test[['MSZoning']])
test[['MSZoning']] = imputer3.transform(test[['MSZoning']])
```

```python
imputer2 = SimpleImputer(strategy='median')

train[['LotFrontage']] = imputer2.fit_transform(train[['LotFrontage']])
test[['LotFrontage']] = imputer2.transform(test[['LotFrontage']])
```

```python
train['BsmtQual'] = train['BsmtQual'].fillna('None')
test['BsmtQual'] = test['BsmtQual'].fillna('None')

train['BsmtCond'] = train['BsmtCond'].fillna('None')
test['BsmtCond'] = test['BsmtCond'].fillna('None')

train['BsmtExposure'] = train['BsmtExposure'].fillna('None')
test['BsmtExposure'] = test['BsmtExposure'].fillna('None')

train['BsmtFinType1'] = train['BsmtFinType1'].fillna('None')
test['BsmtFinType1'] = test['BsmtFinType1'].fillna('None')

train['BsmtFinType2'] = train['BsmtFinType2'].fillna('None')
test['BsmtFinType2'] = test['BsmtFinType2'].fillna('None')

train['FireplaceQu'] = train['FireplaceQu'].fillna("None")
test['FireplaceQu'] = test['FireplaceQu'].fillna("None")

train['GarageType'] = train['GarageType'].fillna('None')
test['GarageType'] = test['GarageType'].fillna('None')

train['GarageFinish'] = train['GarageFinish'].fillna('None')
test['GarageFinish'] = test['GarageFinish'].fillna('None')

train['GarageQual'] = train['GarageQual'].fillna('None')
test['GarageQual'] = test['GarageQual'].fillna('None')

train['GarageCond'] = train['GarageCond'].fillna('None')
test['GarageCond'] = test['GarageCond'].fillna('None')
```

```python

test["GarageCars"] = test["GarageCars"].fillna(train["GarageCars"].median())
```

```python

test["GarageArea"] = test["GarageArea"].fillna(train["GarageArea"].median())
```

```python
imputer3 = SimpleImputer(strategy='most_frequent')


test[['MSZoning']] = imputer3.fit_transform(test[['MSZoning']])
test[['MSZoning']] = imputer3.transform(test[['MSZoning']])

train["GarageYrBlt"] = train["GarageYrBlt"].fillna(0)
test["GarageYrBlt"] = test["GarageYrBlt"].fillna(0)
```
