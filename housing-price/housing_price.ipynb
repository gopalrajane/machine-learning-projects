# Feature Transformation using PowerTransformer (Yeo-Johnson)

This step is used to make data more Gaussian (normal-like), which often improves model performance.

# What is happening here?

We are applying Power Transformation on the dataset to reduce skewness and make features more normally distributed.

# Why PowerTransformer is used?

Many real-world features (like house prices, income, crime rate) are skewed.
This can negatively affect machine learning models.

PowerTransformer helps to:

Reduce skewness
Stabilize variance
Make data closer to normal distribution
Improve model accuracy (especially linear models)

# Why Yeo-Johnson method?

Works with both positive and negative values
Unlike Box-Cox, it does not require strictly positive data
More flexible for real-world datasets

```python
from sklearn.preprocessing import PowerTransformer
import pandas as pd

trf = PowerTransformer(method='yeo-johnson')

x_train_transformed = trf.fit_transform(X_train)
x_test_transformed = trf.transform(X_test)

# convert to DataFrame
x_train_transformed = pd.DataFrame(x_train_transformed, columns=X_train.columns)
x_test_transformed = pd.DataFrame(x_test_transformed, columns=X_test.columns)
```

```python
def qq_plot(feature):
    plt.figure(figsize=(12,4))

    plt.subplot(121)
    sns.histplot(x_train_transformed[feature])
    plt.title(f"{feature} Histogram")

    plt.subplot(122)
    stats.probplot(x_train_transformed[feature], dist='norm', plot=plt)
    plt.title(f"{feature} QQ Plot")

    plt.show()
```

```python
qq_plot('NOX')
```

# What is happening here?

We are building two models:

Model 1 (clf) → trained on original data

🔹 Why Linear Regression?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error

clf = LinearRegression()
clf2 = LinearRegression()

clf.fit(X_train,y_train)
clf2.fit(x_train_transformed,y_train)

y_pred = clf.predict(X_test)
y_pred2 = clf2.predict(x_test_transformed)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse1 = mean_squared_error(y_test, y_pred2)
r21 = r2_score(y_test, y_pred2)

print("MSE:", mse)
print("R2 Score:", r2)


print("MSE:", mse1)
print("R2 Score:", r21)
```

# Hyperparameter Tuning using GridSearchCV

This step is used to find the best combination of parameters for the Random Forest model to improve performance.

# Why Hyperparameter Tuning?

Machine learning models have settings called hyperparameters (like number of trees, depth, etc.).
These are not learned from data — we set them manually.

Tuning helps to:

Improve accuracy
Reduce overfitting
Find optimal model configuration

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators' : [200, 300, 400],
    'max_depth' : [8, 10, 15]
    # 'min_sample_split' : [2, 5, 10],
    # 'min_sample_leaf' : [1, 2, 4]
} 

grid = GridSearchCV(RandomForestRegressor(random_state=40),
                    params, cv=5, scoring='r2')

grid.fit(x_train_transformed, y_train)

print(grid.best_params_)
```

# Log Transformation of Target Variable (Important ML Trick)

This step applies a log transformation on the target variable (y) to improve model performance, especially when the target is skewed.

```python

y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

model.fit(x_train_transformed, y_train_log)

preds = model.predict(x_test_transformed)
preds = np.expm1(preds)


r2_score(y_test, preds)
```
